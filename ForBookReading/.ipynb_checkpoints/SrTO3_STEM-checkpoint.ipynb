```python
import numpy as np
import matplotlib.pyplot as plt
import warnings

import abtem
import ase

from abtem import show_atoms
from ase.visualize import view

warnings.simplefilter('ignore')
```

## Atom configration ==>

```python
sto = ase.io.read("SrTiO3-cub.cif")
view(sto, viewer='ngl')
# Red : oygen , Silver : titanium, Green : Sr
```

```python
sto_2_2_6 = sto * (2, 2, 6)
view(sto_2_2_6, viewer='ngl')
```

## <== Atom configration
## FrozenPhonon ==>

#### argument in FrozenPhonon function
##### parameters for displacement
***num_configs*** : sample number of displaced(Vibrated) atoms   
***sigmas*** : standard deviation of displacement    
***seed*** : for random number generation 

```python
frozen_phonons = abtem.FrozenPhonons(
    sto_2_2_6,
    num_configs=10,
    sigmas=0.1,
    seed=100)
```

##### augument in Potential function 
##### parameters for potential
***sampling*** : spacing of the xy samples of the potential  
***slice_thickness*** : nm  
***projection*** : how those slices are calculated 

```python
potential_frozen = abtem.Potential(
    frozen_phonons,
    sampling=0.05,
    slice_thickness=1,
    projection = "finite"
    )
```

#### visualize sum of potentials at each slices

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
abtem.show_atoms(sto_2_2_6, ax=ax1, legend=True)
visualization = potential_frozen.show(ax=ax2);
```

### <== Frozen Phonon
### Convergent beam==>

#### augument in Probe function
##### parameter for probe wave function
***energy*** : acceleration voltage  
***semiangle_cutoff*** : semi-angle of the electron beam cone  
***Cs*** : Corrector-Spherical Aberration  
***defocus*** : scherzer(**Error occurring**) 

```python
probe = abtem.Probe(energy=80e3, semiangle_cutoff=25, Cs=10e4)
probe.grid.match(potential_frozen)

print(f"defocus = {probe.aberrations.defocus} Å")      # デフォーカス
print(f"FWHM = {probe.profiles().width().compute()} Å")# プローブ径の半値全幅 
```

#### visualize the profile of the probe

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
probe.show(ax=ax1)
probe.profiles().show(ax=ax2);
```

### <== Convergent beam
### ==> Scan and detect

#### augument in GridScan

***start, end*** : region designation  
***nyquist_sampling*** : scan at the Nyquist frequency

```python
grid_scan = abtem.GridScan(
    start=(0, 0),
    end=(1, 1),
    sampling=probe.aperture.nyquist_sampling,
    fractional=True,
    potential=potential_frozen,
)

fig, ax = abtem.show_atoms(sto_2_2_6)

grid_scan.add_to_plot(ax)
```

#### FlexibleAnnularDetector : マルチスライスしたあとでdetector angleを指定できる
#### grid_scan : 2次元平面を均一にsweepします

```python
detector = abtem.FlexibleAnnularDetector()
flexible_measurement = probe.scan(potential_frozen, scan=grid_scan, detectors=detector)
flexible_measurement.compute()

haadf_measurement = flexible_measurement.integrate_radial(90, 200)
```

#### visualize ADF-STEM image before postprocess

```python
haadf_measurement.show(
    explode=True,
    figsize=(14, 5),
    cbar=True,
);
```

### <== Scan and Detect
### Postprocess(ピクセルの平滑化) ==>

#### augument in interpolate function
***sampling*** : the number of interpolation in one pixel

```python
interpolated_measurements = haadf_measurement.interpolate(0.05)
```

#### augument in gaussian_filter function
***standard deviation*** : standard deviation of gauss func

```python
filtered_measurements = interpolated_measurements.gaussian_filter(0.3)
```

#### visualize postprocessed ADF-STEM image

```python
filtered_measurements.show(
    explode=True,
    figsize=(14, 5),
    cbar=True,
);
```
