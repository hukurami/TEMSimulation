## <== Atom configration
## FrozenPhonon ==>

#### parameters for displacement
***num_configs*** : sample number of displaced(Vibrated) atoms   
***sigmas*** : standard deviation of displacement    
***seed*** : for random number generation  

```python
frozen_phonons = abtem.FrozenPhonons(sto_5_5_40, num_configs=10, sigmas=0.1, seed=100)
```

#### parameters for potential
gpts : number of grid points describing the wave functions  
slice_thickness : nm 

```python
potential_phonon = abtem.Potential(frozen_phonons, gpts=432, slice_thickness=2)
```

#### parameter for electron wave function
energy : acceleration voltage

```python
initial_waves = abtem.PlaneWave(energy=150e3)
```

```python
exit_waves_phonon = initial_waves.multislice(potential_phonon)
```

```python
PhononInt = exit_waves_phonon.intensity().mean(0)
```

```python
PhononInt.show(figsize=(10,6), cbar=True, cmap='viridis')
```
